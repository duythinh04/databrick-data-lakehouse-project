```python
query = """
    SELECT 
        ROW_NUMBER() OVER(ORDER BY pi.product_id) as product_key,
        pi.product_id,
        pi.product_code_key,
        pi.product_name,
         CASE 
            WHEN pc.category_id IS NULL THEN 'Uncategorized'
            ELSE pc.category_id
        END AS category_id,
        CASE 
            WHEN pc.sub_category IS NULL THEN 'Uncategorized'
            ELSE pc.sub_category
        END AS sub_category,
        pi.product_cost,
        pi.product_line,
        pc.maintenance,
        pi.product_start_date,
        pi.product_end_date
    FROM data_lakehouse_project.silver.silver_prd_info pi
    LEFT JOIN data_lakehouse_project.silver.silver_px_cat_g1v2 pc
        ON pi.product_category = pc.category_id
"""

df = spark.sql(query)
```

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("data_lakehouse_project.gold.gold_dim_product")
)
```
