```python
query = """
    SELECT 
        ROW_NUMBER() OVER(ORDER BY sd.sales_order_number) as sales_order_id,
        sd.sales_order_number as order_number,
        gp.product_key,
        dc.customer_key,
        sd.sales_order_date as order_date,
        sd.sales_ship_date as shipping_date,
        sd.sales_due_date as due_date,
        sd.sales_sum_sales as sales_sum,
        sd.sales_quantity as sales_amount,
        sd.sales_price as price
    FROM data_lakehouse_project.silver.silver_sale_details sd
    LEFT JOIN data_lakehouse_project.gold.gold_dim_product gp
        ON sd.sales_product_key = gp.product_code_key
    LEFT JOIN data_lakehouse_project.gold.gold_dim_customer dc
        ON sd.sales_customer_id = dc.customer_id
"""


df = spark.sql(query)
```

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("data_lakehouse_project.gold.gold_fact_sales")
)
```
