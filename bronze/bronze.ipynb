#READ FILE FROM CSV & WRITE TO TABLE

```python
from bronze_config import INGESTION_CONFIG
for item in INGESTION_CONFIG:
    print(f"Ingesting {item["path"]} to table {item["table"]}")

    df= (
        spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(item["path"])
    )
    (
        df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(item["table"])
    )
```
