```python
import pyspark.sql.functions as F
from pyspark.sql.functions import col, trim
from pyspark.sql.types import StringType
```

```python
df = spark.table("data_lakehouse_project.bronze.bronze_sales_details")
```

```python
RENAME_MAP = {
    "sls_ord_num": "sales_order_number",
    "sls_prd_key": "sales_product_key",
    "sls_cust_id": "sales_customer_id",
    "sls_order_dt": "sales_order_date",
    "sls_ship_dt": "sales_ship_date",
    "sls_due_dt" : "sales_due_date",
    "sls_sales": "sales_sum_sales",
    "sls_quantity": "sales_quantity",
    "sls_price": "sales_price"
}
```

#renaming header name

```python
#Convert int to Date type
df = (
    df.withColumn("sls_order_dt", F.expr("try_to_date(cast(sls_order_dt as string), 'yyyyMMdd')"))
      .withColumn("sls_ship_dt", F.expr("try_to_date(cast(sls_ship_dt as string), 'yyyyMMdd')"))
      .withColumn("sls_due_dt", F.expr("try_to_date(cast(sls_due_dt as string), 'yyyyMMdd')"))
)
```

```python
for old_name,new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name,new_name)
```

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .saveAsTable("data_lakehouse_project.silver.silver_sale_details")
)
```
